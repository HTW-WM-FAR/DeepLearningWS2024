```python
import matplotlib.pyplot as plt
import pandas as pd

# Load results from TimeLLM, LSTM, and ARIMA
results_timellm = pd.read_csv("TIMELLM_RESULTS_PER_STOCK/results_timellm.csv")  
results_lstm = pd.read_csv("LSTM_RESULTS_PER_STOCK/results_lstm.csv")  
results_arima = pd.read_csv("ARIMA_RESULTS_PER_STOCK/arima_comparison_results.csv")  

# Get unique stock tickers
unique_stocks = set(results_timellm["unique_id"].unique()) | set(results_lstm["unique_id"].unique()) | set(results_arima["unique_id"].unique())

# Iterate over each stock and plot the graphs
for stock in unique_stocks:
    # Filter data for the current stock
    stock_llm = results_timellm[results_timellm["unique_id"] == stock]
    stock_lstm = results_lstm[results_lstm["unique_id"] == stock]
    stock_arima = results_arima[results_arima["unique_id"] == stock]

    # Plot MAE vs. Interval
    plt.figure(figsize=(8, 5))
    plt.plot(stock_llm["Interval"], stock_llm["MAE"], marker='o', linestyle='-', label="TimeLLM MAE", color="blue")
    plt.plot(stock_lstm["Interval"], stock_lstm["MAE"], marker='s', linestyle='--', label="LSTM MAE", color="green")
    plt.plot(stock_arima["Interval"], stock_arima["MAE"], marker='^', linestyle='-.', label="ARIMA MAE", color="red")
    plt.xlabel("Interval")
    plt.ylabel("MAE")
    plt.title(f"MAE Comparison for {stock}")
    plt.grid()
    plt.legend()
    plt.show()

    # Plot RMSE vs. Interval
    plt.figure(figsize=(8, 5))
    plt.plot(stock_llm["Interval"], stock_llm["RMSE"], marker='o', linestyle='-', label="TimeLLM RMSE", color="blue")
    plt.plot(stock_lstm["Interval"], stock_lstm["RMSE"], marker='s', linestyle='--', label="LSTM RMSE", color="green")
    plt.plot(stock_arima["Interval"], stock_arima["RMSE"], marker='^', linestyle='-.', label="ARIMA RMSE", color="red")
    plt.xlabel("Interval")
    plt.ylabel("RMSE")
    plt.title(f"RMSE Comparison for {stock}")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load results from CSV files
results_timellm = pd.read_csv("TIMELLM_RESULTS_PER_STOCK/results_timellm.csv")  
results_lstm = pd.read_csv("LSTM_RESULTS_PER_STOCK/results_lstm.csv")  
results_arima = pd.read_csv("ARIMA_RESULTS_PER_STOCK/arima_comparison_results.csv")  

# Get unique stock tickers
unique_stocks = set(results_timellm["unique_id"].unique()) | set(results_lstm["unique_id"].unique()) | set(results_arima["unique_id"].unique())

# Function to plot RMAE for each model
plt.figure(figsize=(10, 6))
for stock in unique_stocks:
    stock_llm = results_timellm[results_timellm["unique_id"] == stock]
    plt.plot(stock_llm["Interval"], stock_llm["RMAE"], marker='o', linestyle='-', label=stock)
plt.xlabel("Interval")
plt.ylabel("RMAE (%)")
plt.title("RMAE Comparison for TimeLLM")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
for stock in unique_stocks:
    stock_lstm = results_lstm[results_lstm["unique_id"] == stock]
    plt.plot(stock_lstm["Interval"], stock_lstm["RMAE"], marker='s', linestyle='--', label=stock)
plt.xlabel("Interval")
plt.ylabel("RMAE (%)")
plt.title("RMAE Comparison for LSTM")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
for stock in unique_stocks:
    stock_arima = results_arima[results_arima["unique_id"] == stock]
    plt.plot(stock_arima["Interval"], stock_arima["RMAE"], marker='^', linestyle='-.', label=stock)
plt.xlabel("Interval")
plt.ylabel("RMAE (%)")
plt.title("RMAE Comparison for ARIMA")
plt.legend()
plt.grid()
plt.show()

# Function to plot RMSE for each model
plt.figure(figsize=(10, 6))
for stock in unique_stocks:
    stock_llm = results_timellm[results_timellm["unique_id"] == stock]
    plt.plot(stock_llm["Interval"], stock_llm["RMSE"], marker='o', linestyle='-', label=stock)
plt.xlabel("Interval")
plt.ylabel("RMSE")
plt.title("RMSE Comparison for TimeLLM")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
for stock in unique_stocks:
    stock_lstm = results_lstm[results_lstm["unique_id"] == stock]
    plt.plot(stock_lstm["Interval"], stock_lstm["RMSE"], marker='s', linestyle='--', label=stock)
plt.xlabel("Interval")
plt.ylabel("RMSE")
plt.title("RMSE Comparison for LSTM")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
for stock in unique_stocks:
    stock_arima = results_arima[results_arima["unique_id"] == stock]
    plt.plot(stock_arima["Interval"], stock_arima["RMSE"], marker='^', linestyle='-.', label=stock)
plt.xlabel("Interval")
plt.ylabel("RMSE")
plt.title("RMSE Comparison for ARIMA")
plt.legend()
plt.grid()
plt.show()
```
